```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv("USA_cars_datasets.csv")
data=data.head(30)
plt.figure(figsize=(10, 6))
sns.barplot(x='year', y='price', hue='brand', data=data,ci=None)
plt.title('Line Plot of Price Over Years by Brand')
plt.xlabel('Year')
plt.ylabel('Price')
```

```python
data['mileage'].skew()
```

```python
sns.distplot(data['mileage'],hist=True)
```

```python
sq_mileage=np.sqrt(data['mileage'])
```

```python
sq_mileage.skew()
```

```python
sns.distplot(sq_mileage,hist=True)
```

```python
cbrt_mileage=np.cbrt(sq_mileage)
```

```python
cbrt_mileage.skew()
```

```python
sns.distplot(cbrt_mileage,hist=True)
```
