```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
```

```python
data=pd.read_csv("USA_cars_datasets.csv")
data=data.head(300)
```

```python
data['mileage'].skew
```

```python
sns.distplot(data['mileage'],hist=True)
```

```python
sq_mileage=np.sqrt(data['mileage'])
```

```python
sq_mileage.skew()
```

```python
sns.distplot(sq_mileage,hist=True)
```

```python
cbrt_mileage=np.cbrt(data['mileage'])
```

```python
cbrt_mileage.skew()
```

```python
sns.distplot(cbrt_mileage,hist=True)
```
