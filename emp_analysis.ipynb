```python
import pandas as pd 
import matplotlib.pyplot as plt 
import random

```

```python
n=4
employee_data = {
    'Employee Code': [i + 1001 for i in range(n)],
    'Employee Name': ['Employee' + str(i + 1) for i in range(n)],
    'Employee Designation': ['Designation' + str(i + 1) for i in range(n)],
    'Employee Salary': [random.randint(4000, 10000) for _ in range(n)]
}
df = pd.DataFrame(employee_data)
df.to_excel('employee_data.xlsx', index=False)
print(df)
```

```python
# Display name of employees whose salary is greater than 5000
filtered_salary = df[df['Employee Salary'] > 5000]
print(filtered_salary)
```

```python
# Display emp. Name, designation and salary
data_fil=df[['Employee Name','Employee Designation','Employee Salary']]
data_fil
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Employee Name'], df['Employee Salary'], color='blue')
plt.xlabel('Employee Name')
plt.ylabel('Employee Salary')
plt.title('Employee Salary Distribution')
```
