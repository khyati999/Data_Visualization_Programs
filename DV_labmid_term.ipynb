```python
import seaborn as sns
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"retail_sales_dataset.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.describe()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Total Amount', marker='o', color='red',ci=None)
plt.xlabel('Date')
plt.ylabel('Total Amount (Sales)')
plt.legend(title="Retail Sales over time")
plt.show()
```

```python
df=df.head(300)
df['Total Amount'].skew()
```

```python
sns.distplot(df['Total Amount'],hist=True)
```

```python
sq_amount=np.sqrt(df['Total Amount'])
sq_amount.skew()
```

```python
sns.distplot(sq_amount,hist=True)
```

```python
cube_root_amount=np.cbrt(sq_amount)
```

```python
cube_root_amount.skew()
```

```python
sns.distplot(cube_root_amount,hist=True)
```
