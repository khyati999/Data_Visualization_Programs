```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"retail_sales_dataset.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.describe()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Total Amount', marker='o', color='red',ci=None)
#plt.title('Retail Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Amount (Sales)')
plt.legend(title="Retail Sales over time")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Date', y='Total Amount', data=df, color='green', marker='o')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Amount (Sales)')
plt.show()
```

```python
sns.heatmap(df[["Total Amount","Date"]].corr(), annot=True)
plt.title('Sales Over Time')
plt.show()
```
