```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
dataset=pd.read_csv(r"Bengaluru_House_Data.csv")
```

```python
column=dataset['price']
mean=column.mean()
median=column.median()
stdeviation=column.std()
plt.hist(column, color='blue', edgecolor='black')
plt.title("Histogram for magnitude")
plt.show()
```

```python
print("Skewness")
sk=3*(mean-median)/stdeviation
print(sk)
```

```python
square_root=np.sqrt(dataset['price'])
square_root
```

```python

meansqrt=square_root.mean()
mediansqrt=square_root.median()
stdeviationsqrt=square_root.std()
plt.hist(square_root, color='blue', edgecolor='black')
plt.title("Histogram for magnitude")
plt.show()
```

```python
print("Skewness")
sk1=3*(meansqrt-mediansqrt)/stdeviationsqrt
print(sk1)
```

```python
log=np.log10(dataset['price'])
meanlog=log.mean()
medianlog=log.median()
stdeviationlog=log.std()
plt.hist(log, color='blue', edgecolor='black')
plt.title("Histogram for magnitude")
plt.show()
print("Skewness")
sklog=3*(meanlog-medianlog)/stdeviationlog
print(sklog)
```
